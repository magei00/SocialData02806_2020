```python
import numpy as np
import pandas as pd
import folium
```

```python
country_geo = 'datasets/all_countries.geo.json'
```

```python
covid_data = pd.read_csv('datasets/COVID-19-worldwide.csv')
```

```python
covid_data.head()
```

```python
present_day = covid_data[covid_data['dateRep'] == '4/27/2020']
present_day.to_csv('present_day.csv')
```

```python
present_day.head()
```

```python
m = folium.Map(location=[48, -102], zoom_start=3)
folium.Choropleth(
    geo_data=country_geo,
    name='choropleth',
    data=present_day,
    columns=['countryterritoryCode', 'deaths'],
    key_on='properties.iso_a3',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='deaths'
).add_to(m)
```
